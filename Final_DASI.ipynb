```python
# Pilih kolom numerik
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Fungsi deteksi outlier pakai IQR
outlier_summary = {}
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_summary[col] = len(outliers)

# Tabel ringkasan jumlah outlier per kolom
outlier_df = pd.DataFrame.from_dict(outlier_summary, orient='index', columns=['Jumlah Outlier'])
outlier_df = outlier_df.sort_values(by='Jumlah Outlier', ascending=False)
print("\n=== JUMLAH OUTLIER PER KOLOM ===")
display(outlier_df)
```

```python
# Handling Missing Values


# Tangani kolom delay (banyak missing tapi maknanya = 0 delay)
delay_cols = ['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
df[delay_cols] = df[delay_cols].fillna(0)

# Tangani kolom numerik dengan missing kecil menggunakan median
cols_with_small_missing = ['ActualElapsedTime', 'AirTime', 'ArrDelay', 'ArrTime', 'TaxiIn', 'TaxiOut', 'CRSElapsedTime']
for col in cols_with_small_missing:
    df[col] = df[col].fillna(df[col].median())

# Tangani TailNum (identitas pesawat) dengan mode
df['TailNum'] = df['TailNum'].fillna(df['TailNum'].mode()[0])

# Cek ulang
print("=== Missing values setelah ditangani ===")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
# Cek kembali Missing Values
# Cek ulang apakah masih ada missing value
print(df.isnull().sum())
```

```python
# Handling Outliers

from scipy.stats.mstats import winsorize

# Pilih kolom yang sensitif terhadap outlier (delay dan waktu)
outlier_cols = ['DepDelay', 'ArrDelay', 'CarrierDelay', 'WeatherDelay', 
                'NASDelay', 'SecurityDelay', 'LateAircraftDelay', 'TaxiIn', 'TaxiOut']

# Terapkan winsorization (1% bawah dan 99% atas)
for col in outlier_cols:
    df[col] = winsorize(df[col], limits=[0.01, 0.01])

print("Outlier telah ditangani dengan Winsorization.")
```

```python
from sklearn.ensemble import RandomForestRegressor

# Pilih fitur penting secara manual
selected_cols = ['DepDelay', 'Distance', 'TaxiOut', 
                 'CarrierDelay', 'WeatherDelay', 'NASDelay', 'LateAircraftDelay']

X = df_sample[selected_cols]
y = df_sample['ArrDelay']

rf = RandomForestRegressor(n_estimators=30, random_state=42, n_jobs=-1)
rf.fit(X, y)

importances = pd.Series(rf.feature_importances_, index=selected_cols).sort_values(ascending=False)
print("\nFeature Importance:")
print(importances)

plt.figure(figsize=(8,5))
importances.plot(kind='bar', color='orange')
plt.title('Feature Importance terhadap ArrDelay (Sample 10.000 Data)')
plt.ylabel('Tingkat Kepentingan Fitur')
plt.show()
```

```python
# Uji Parametik Pearson Correlation
from scipy.stats import pearsonr

# Ambil dua variabel utama
x = df['DepDelay']
y = df['ArrDelay']

# Hapus missing value agar analisis valid
data_clean = df[['DepDelay', 'ArrDelay']].dropna()

# Uji Pearson correlation
r_value, p_value = pearsonr(data_clean['DepDelay'], data_clean['ArrDelay'])

print(f"Pearson Correlation Coefficient (r): {r_value:.3f}")
print(f"P-value: {p_value:.5f}")
```

```python
import numpy as np
import scipy.stats as st

n = len(data_clean)
stderr = 1.0 / np.sqrt(n - 3)
delta = 1.96 * stderr  # 95% CI
lower = np.tanh(np.arctanh(r_value) - delta)
upper = np.tanh(np.arctanh(r_value) + delta)
print(f"95% Confidence Interval for r: [{lower:.3f}, {upper:.3f}]")
```

```python
# Kesimpulan Uji Parametrik:

Terdapat hubungan linear yang sangat kuat dan signifikan antara keterlambatan keberangkatan dan keterlambatan kedatangan.
Semakin besar DepDelay, semakin besar pula ArrDelay.
```

```python
# Uji Non-Parametrik – Spearman Rank Correlation
from scipy.stats import spearmanr

rho, pval = spearmanr(data_clean['DepDelay'], data_clean['ArrDelay'])

print(f"Spearman Correlation (rho): {rho:.3f}")
print(f"P-value: {pval:.5f}")
```
